```python
cal = {
    'Day 1':1750,
    'Day 2':2100,
    'Day 3':300
}
t = pd.Series(cal)

print(t.loc['Day 1'])
print(t.iloc[1])
print(t)
t.loc['Day 1'] = 1000
print(t)

print(cal[cal < 2000])
```

```python
#DataFram : A tabular data structure with rows and columns. 2-D
import pandas as pd
dat = {
    "Name":["bob","cab","mathe"],
    "Age":[20,30,40]
}
df = pd.DataFrame(dat,index=['Enpl 1','Emp 2','Emp 3'])
print(df.loc['Enpl 1'])
print(df.iloc[1])
print(df)
#ADD NEW COLUMN
df['Job'] = ['Engineer','N/A','Manager']
print(df)
#Add New Row
new_row = pd.DataFrame([{'Name':'Sandy','Age':28,'Job':'teacher'},{'Name':'San','Age':20,'Job':'Painter'}],index=['Emp 5','Emp 6'])
print(new_row)
df = pd.concat([df,new_row],ignore_index=True)
print(df)
```

#Aggregrate

```python
#Reduces a set of values into a single summary value - groupby()
df = pd.read_csv('Mall_Customers.csv')
print(df.mean(numeric_only=True))
print("Sum--------------")
print(df.sum(numeric_only=True))
print("min---------------")
print(df.min(numeric_only=True))
print("Count",df.count())
```

```python
#Single column
print(df['Age'].mean(numeric_only=True))
print("Sum--------------")
print(df['Age'].sum(numeric_only=True))
print("min---------------")
print(df['Age'].min(numeric_only=True))
print("Count",df.count())
```

```python
g= df.groupby('Age')
```

```python
print(g['Age'].mean())

```

```python
print(g['Age'].count())
```
